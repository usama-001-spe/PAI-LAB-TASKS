```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
```

```python
data=pd.read_csv('train.csv')
df=pd.DataFrame(data)
df
```

```python
df.isnull().sum()
```

```python
traindf2=df.copy()
```

```python
traindf2['Age'].fillna(traindf2['Age'].mean(), inplace=True)

```

```python
encoder=LabelEncoder()
traindf2['VIP']=encoder.fit_transform(traindf2['VIP'])
```

```python
traindf2['VIP'].isnull
```

```python
encoder=LabelEncoder()
traindf2['Transported']=encoder.fit_transform(traindf2['Transported'])
```

```python
x=traindf2[['Age','VIP']]
y=traindf2['Transported']
```

```python
model=RandomForestClassifier()
model.fit(x,y)
```

```python
data2=pd.read_csv('test.csv')
df2=pd.DataFrame(data2)
df2
```

```python
df2.isnull().sum()
```

```python
testdf2=traindf2.copy()
```

```python
testdf2['Age'].fillna(testdf2['Age'].mean(), inplace=True)
```

```python
encoder=LabelEncoder()
testdf2['VIP']=encoder.fit_transform(testdf2['VIP'])
```

```python
testx=testdf2[['Age','VIP']]
```

```python
predicted=model.predict(testx)
```

```python
submission=pd.DataFrame({'PassengerID':testdf2['PassengerId'],'Transported':predicted})
```

```python
submission.to_csv('submission.csv',index=False)
```
